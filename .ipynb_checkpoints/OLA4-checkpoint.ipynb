```python
import numpy as np
import pandas as pd
import random as ran
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import cv2
import os
from sklearn.model_selection import train_test_split
import tensorflow as tf
import keras
```

```python
train_path = './CNN/train/'
train_data = []

test_path = './CNN/test/'
test_data = []

for filename in os.listdir(train_path):
    if filename.startswith("cat") or filename.startswith("dog"):
        label = filename.split('.')[0]
        if (label == "cat"):
            label = 0
        else:
            label = 1
        img = cv2.imread(os.path.join(train_path, filename))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (28, 28)) # Have to make pictures very small because my laptop can't handle it if they are bigger
        train_data.append((img, label))

for filename in os.listdir(test_path):
    img = cv2.imread(os.path.join(test_path, filename))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (28, 28)) # Have to make pictures very small because my laptop can't handle it if they are bigger
    label = filename.split('.')[0]
    test_data.append((img, label))
```

```python
print("Number of training images:", len(train_data))
print("Number of test images:", len(test_data))
```

```python
train_data[0]
```

```python
# Plotting random images from the training set

ran.seed(27)
random_images = ran.sample(train_data, 9)

plt.figure(figsize=(4, 4))
for i, (img, label) in enumerate(random_images):
    plt.subplot(3, 3, i + 1)
    plt.imshow(img)
    if label == 0:
        plt.title("Cat")
    else:
        plt.title("Dog")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Splitting the data up into X for features and y for labels
X_train = []
y_train = []
X_test = []
y_test = []

for feature, label in train_data:
    X_train.append(feature)
    y_train.append(label)

for feature, label in test_data:
    X_test.append(feature)
    y_test.append(label)
    
X_train = np.array(X_train)
y_train = np.array(y_train)

X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
print(y_test)
```

```python
print(y_train)
```

```python
# Splitting the data into validation and training
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
```

```python
# Normalizing pixel values for train_data
X_train = X_train / 255.0
X_val = X_val / 255.0
```

```python
# Normalizing pixel values for test_data
X_test = X_test / 255.0
```

```python
# Train
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_val shape:", X_val.shape)
print("y_val shape:", y_val.shape)
```

```python
# Test
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# One hot-encoding
y_train = y_train.astype(np.int32)
```

```python
y_test = y_test.astype(np.int32)
```

```python
y_train = tf.one_hot(y_train, depth=10)
y_test = tf.one_hot(y_test, depth=10)
```

```python
from keras.models import Sequential
from keras.layers import Input, Conv2D, MaxPool2D, Flatten, Dense, Dropout
from keras.optimizers import Adam
```

```python
# Setting values for the model
input_shape = (28, 28, 3)
classes = 10
epochs = 10
learning_rate = 0.001
batch_size = 256
```

```python
model = Sequential([
    Conv2D(16, (3, 3), activation='relu', padding='same', input_shape=input_shape),
    Conv2D(16, (3, 3), activation='relu', padding='same'),
    MaxPool2D(strides=(2,2)),
    Dropout(0.25),
    
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    MaxPool2D(strides=(2,2)),
    Dropout(0.25),
    
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.25),
    Dense(classes, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=learning_rate), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, batch_size=batch_size, validation_split=0.1, epochs=epochs)
```

```python
accuracy      = history.history['accuracy']
val_accuracy  = history.history['val_accuracy']
loss          = history.history['loss']
val_loss      = history.history['val_loss']
```

```python
plt.plot(accuracy)
plt.plot(val_accuracy)
plt.plot(loss)
plt.plot(val_loss)

plt.title("Model Accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(["accuracy", "val_accuracy", "loss", "val_loss"])
print(history.history.keys())
plt.show()
```

Based on your experience, how would you define ANN and CNN?

A artificial neural network (ANN) is a machine learning model inspired by the way the brain functions. It consists of many artificial neurons organized into layers. The three layers are the input layer, hidden layers, and output layer.

A convolutional neural network (CNN) is a type of artificial neural network architecture that is effective for image recognition and image classification. It comprises several layers including the input layer, convolutional layers, activation layers, pooling layers, fully connected layers, and output layer.

Which hyper-parameters of the CNN architecture have you fine-tuned?

I have adjusted the number of convolutional layers (Conv2D) and the number of filters in each layer. Increasing the number of filters can allow the model to learn more complex patterns but might increase computational cost.

The size of the convolutional kernels (the (3,3) part of Conv2D) makes it so that it can capture different levels of spatial information.

Activation functions (activation='relu') can affect how the model learns and generalizes. Using different activations like tanh or LeakyReLU instead of relu might get different results.

I have adjusted the parameters of the pooling layers (MaxPool2D) like strides which can impact the spatial dimensions of the output feature maps.

I use dropout (Dropout(0.25)) to prevent overfitting.

The number of neurons in the dense layers (Dense(128, activation='relu')) are adjusted based on the complexity and size of the dataset.
